```python
# 1.기본
import numpy as np  # numpy 패키지 가져오기
import matplotlib.pyplot as plt # 시각화 패키지 가져오기
import seaborn as sns # 시각화

# 2.데이터 가져오기
import pandas as pd # csv -> dataframe으로 전환

# 3.통계분석 package
import pingouin as pg
from scipy import stats
import statsmodels.api as sm
```

## 2.데이터 불러오기

### 2.1 데이터 프레임으로 저장
- 원본데이터(csv)를 dataframe 형태로 가져오기(pandas)

```python
rma_df = pd.read_csv('https://raw.githubusercontent.com/leecho-bigdata/statistics-python/main/08_1.RMA.csv', encoding="cp949")
rma_df.head()
```

### 2.2 범주형 변수 처리
- 가변수 처리시 문자로 처리를 해야 변수명 구분이 쉬움

```python
rma_df['시간'].replace({1:'사전점수', 2:'중간점수', 3:'최종점수'}, inplace=True)
rma_df['시간'] = rma_df['시간'].astype('category')

rma_df.head()
```

## 3.기술통계

```python
# 그룹별 기술통계
rma_df.groupby('시간')["점수1"].describe().round(3)
```

```python
# 분석변수가 여러개 일 때
num_feature = ['점수1', '점수2', '점수3']
for num in num_feature:
    print("----", num, "----")
    results = rma_df.groupby("시간")[num].describe().round(3)
    print(results, "\n")
```

```python
sns.catplot(data = rma_df,
            x = "시간",
            y = "점수1",
            hue = "시간",
            kind = "box")
plt.show()
```

## 4.구형성 검정

```python
# 구형성이면 지금까지 분석한 것이 문제 없음
# Mauchly: p = 0.126, p >= 0.05 이므로 귀무가설(H0) 채택 -> 구형성 성립
pg.sphericity(dv = '점수1',
              within = '시간',
              subject = "id",
              data = rma_df)
```

```python
num_feature = ['점수1', '점수2', '점수3']
for num in num_feature:
    print("----", num, "----")
    results = pg.sphericity(dv = num,
                            within = '시간',
                            subject = "id",
                            data = rma_df)
    print(results, "\n")
```

## 5.ANOVA(RM)

### 5.1 구형성일때, 차이가 없을때

```python
# 구형성일때, 차이가 없을때
# ANOVA: p = 0.088, p >= 0.05 이므로 귀무가설(H0) 채택 -> 차이없음
# correction = False
pg.rm_anova(dv = '점수1',
            within = '시간',
            subject = "id",
            correction = False,
            data = rma_df,
            detailed = True).round(3)
```

```python
# 그래프
g = sns.catplot(x = "시간",
                y = "점수1",
                kind = "point",
                data = rma_df)
```

### 5.2 구형성일때, 차이가 있을때

```python
# 구형성일때, 차이가 있을때
# ANOVA: p = 0.000, p < 0.05 이므로 귀무가설(H0) 기각 -> 차이있음
# correction = False
pg.rm_anova(dv = '점수2',
            within = '시간',
            subject = "id",
            correction = False,
            data = rma_df,
            detailed = True).round(3)
```

```python
# post-hoc test(RMA)
pg.pairwise_tests(dv = '점수2',
                  within = '시간',
                  subject = "id",
                  data = rma_df).round(3)
```

```python
# 그래프
sns.catplot(x = "시간",
            y = "점수2",
            kind = "point",
            data = rma_df)
```

### 5.3 구형성이 아닐 때(Greenhouse-Geisser)

```python
# 구형성이 아닐때: Greenhouse-Geisser corrected p-value
# ANOVA: p-GG-corr = 0.000, p < 0.05 이므로 귀무가설(H0) 기각 -> 차이있음
# correction = True
pg.rm_anova(dv = '점수3',
            within = '시간',
            subject = "id",
            correction = True,
            data = rma_df,
            detailed = True).round(3)
```

```python
# post-hoc test(RMA)
pg.pairwise_tests(dv = '점수3',
                  within = '시간',
                  subject = "id",
                  data = rma_df).round(3)
```

```python
# 그래프
g = sns.catplot(x = "시간",
                y = "점수3",
                kind = "point",
                data = rma_df)
```

## 7.Friedman(비모수)

```python
# friedman
pg.friedman(dv = '점수3',
            within = '시간',
            subject = "id",
            data = rma_df).round(3)
```

```python
# Non-parametric pairwise paired test (wilcoxon)
pg.pairwise_tests(dv = '점수3',
                  within = '시간',
                  subject = "id",
                  parametric = False,
                  data = rma_df).round(3)
```
